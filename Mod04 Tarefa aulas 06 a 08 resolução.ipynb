####  2) Melhorando a interação com o API
Observe dois URLs de consultas diferentes, por exemplo o URL utilizado em aula, e o URL feito no exercício anterior. Compare-os e observe as diferenças.

1. Faça uma função em Python que recebe como argumento o UF da consulta e o tipo de interessado (```'Estado'```ou ```Município```), e que devolve os dados da consulta no formato *DataFrame*.
2. Quantas solicitações para o Estado podem ser consultadas para Minas Gerais com *status* em 'Arquivado por decurso de prazo' estão registradas?
3. Qual é o município da Bahia com mais solicitações deferidas?
4. Salve um arquivo .csv com os dados de solicitações da Bahia, com interessado = 'Estado'

````python
#1) Seu código aqui

def pesquisar_inf():
    uf = input('Informe a sigla do estado de interesse. Ex: sp, mg, rj, ...').upper()
    interesse = input('Informe a entidade da busca, estado ou município.')
    if interesse == 'município':
        url = f'https://apidatalake.tesouro.gov.br/ords/sadipem/tt/pvl?uf={uf}&tipo_interessado=Munic%C3%ADpio'
    elif interesse == 'estado':
        url = f'https://apidatalake.tesouro.gov.br/ords/sadipem/tt/pvl?uf={uf}&tipo_interessado=Estado'
        
    dados = requests.get(url)
    dados_json = dados.json()
    dados = pd.DataFrame(dados_json['items'])
    return dados
````

````python
df_dados.columns
````

````python
# Filtrar dados
df_deferidos = df_dados[df_dados['status'] == 'Deferido']

municipios_agrupados = df_deferidos.groupby('interessado')['interessado'].count()
municipios_agrupados
````

````python
municipio_mais_deferidos = municipios_agrupados.idxmax()
municipio_mais_deferidos
````

````python
num_de_solic_def = municipios_agrupados[municipio_mais_deferidos]
num_de_solic_def
````
